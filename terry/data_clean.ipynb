```python
# load required libraries
import pandas as pd
```

```python
# import file to work on
inventory_df = pd.read_csv('Resources/BM_products.csv')
inventory_df.head()
```

```python
# trim file down to required colums for use
target_invetory_df = inventory_df[['Product Name', 'num_of_inv', 'Volume', 'List Price']]
target_invetory_df.head()
```

```python
# clean data in file for use with speech
target_invetory_df = target_invetory_df.rename(columns={'Product Name': 'product', 'num_of_inv': 'inventory', 'Volume': 'volume', 'List Price': 'price'})
target_invetory_df.head()
```

```python
# output modified file into csv for use with speech bot
target_invetory_df.to_csv('Resources/final_inventory.csv', index= False)
```
